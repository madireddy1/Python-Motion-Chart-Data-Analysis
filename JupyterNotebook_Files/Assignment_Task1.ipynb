# Task 1: Unemployment Rate and House Prices
In the task, you are required to visualise the relationship between the unemployment rate, the house price index,
and the population of different Australian states, and gain insights from how these relations and trends change
over time using motion chart in python.

```python
#import the motion chart and pandas library
from motionchart.motionchart import MotionChart
import pandas as pd
```

```python
#Read the Australian Statewise population data from the file using read csv
erp_state_population_data = pd.read_csv('ERP_by state and gender.csv')
#check the data by displaying the top records using head
erp_state_population_data.head()
```

```python
#First step of data wrangling by formatting our data focussing on the required columns
erp_state_population_data.columns = ['Date', '','','','','','','','','','','','','','','','','','','NSW','VIC','QLD','SA'
,'WA','TAS','NT','ACT','']
#check the data by displaying the top records
erp_state_population_data.head()
```

```python
#Here we melt the population data and give respective column names
erp_state_population_data=pd.melt(erp_state_population_data,id_vars=['Date'],value_vars=['VIC','QLD','SA','WA','TAS','NT','ACT','NSW'],var_name='States')
erp_state_population_data.rename(columns = {'value':'Population'}, inplace = True)
#check the data by displaying the top records
erp_state_population_data.head()
```

```python
#Take the employment data from the file using read excel and from the respective sheet from excel
employment_data=pd.read_excel("SA4 Time Series - October 2016.xls",sheet_name='Time Series')
#check the data by displaying the top records
employment_data.head()
```

```python
#set up the data format as we need as the process of data wrangling
employment_data.columns = ['State', 'Date', 'employmentRate','unemploymentRate','participiationRate']
employment_data.columns=employment_data.columns.str.strip()
employment_data=employment_data[['State','Date','unemploymentRate']]
# check the data by displaying the top records
employment_data.head()
```

```python
#take house index data from the file using read csv
houseindex_data=pd.read_csv("House Price Index.csv")
#check the data by displaying the top records
houseindex_data.head()
```

```python
#set up the data format as we need
houseindex_data.columns = ['Date', 'NSW','VIC','QLD','SA','WA','TAS','NT','ACT']
#check the data by displaying the top records
houseindex_data.head()
```

```python
#melt the house index data
houseindex_data=pd.melt(houseindex_data,id_vars=['Date'],value_vars=['NSW','VIC','QLD','SA','WA','TAS','NT','ACT'],var_name='States')
houseindex_data.rename(columns = {'value':'HousePriceIndex'}, inplace = True)
houseindex_data.head()
```

```python
# changing format of date here for all the data sets
employment_data.Date=pd.to_datetime(employment_data.Date)
employment_data['Date']=employment_data['Date'].dt.strftime('%d/%m/%y')
houseindex_data.Date=pd.to_datetime(houseindex_data.Date)
houseindex_data['Date']=houseindex_data['Date'].dt.strftime('%d/%m/%y')
erp_state_population_data.Date=pd.to_datetime(erp_state_population_data.Date)
erp_state_population_data['Date']=erp_state_population_data['Date'].dt.strftime('%d/%m/%y')
```

```python
#check the date format by displaying the top the records
erp_state_population_data.head()
```

```python
#check the date format by displaying the top the records
houseindex_data.head()
```

```python
#check the date format by displaying the top the records
erp_state_population_data.head()
```

```python
#merging the employment data and the house index data
merge_data= pd.merge(employment_data,houseindex_data,left_on=['Date','State'],right_on=['Date','States'])
#check the data by displaying the top records
merge_data.head()
```

```python
#now merging the erp_state_population data with the other 2 data sets from merged data
final_data=pd.merge(merge_data,erp_state_population_data,left_on=['Date','State'],right_on=['Date','States'])
#check the final data by displaying the top records
final_data.head()
```

```python
#The final data set is removed with the repetitive columns and formatted properly
final_data = final_data[['State','Date','unemploymentRate','HousePriceIndex','Population']]
#check the final data by displaying the top records
final_data.head()
```

```python
#check the final data using tail command
final_data.tail()
```

```python
#creating motion chart
motionChart = MotionChart(df = final_data, key='Date', x='unemploymentRate', y='HousePriceIndex', xscale='linear', yscale='linear', size='Population', color='State',category='State')
motionChart.to_notebook()
```
