```python
import pandas as pd
```

```python
maxHPI = pd.read_csv('Task1Results.csv')
```

```python
maxHPIResult = maxHPI.groupby('Date')['HousePriceIndex'].max()
```

```python
maxHPIResult = maxHPIResult.to_frame()
maxHPIResult['Date']=maxHPIResult.index
maxHPIResult.reset_index(drop=True, inplace=True)
maxHPIResult.head()
```

```python
FinalMaxHPI=pd.merge(maxHPI,maxHPIResult,left_on=['Date','HousePriceIndex'],right_on=['Date','HousePriceIndex'])
```

```python
FinalMaxHPI.sort_values('State')
FinalMaxHPI[['State','Date', 'HousePriceIndex']]
```
