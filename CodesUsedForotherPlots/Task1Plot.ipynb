# Justification graphs as shown in the PDF answers.



```python
#NT UR vs Date

import pandas as pds
import matplotlib.pyplot as mpyplt

NTR= pds.read_excel("Northern.xlsx")
mpyplt.plot(NTR["Date"],NTR["Unemployement"], label = "NT")
mpyplt.legend(loc='upper center', bbox_to_anchor=(1.0, 1.1),
          ncol=3, fancybox=True, shadow=True)
mpyplt.xlabel('Date')
mpyplt.ylabel('Unemployement')
mpyplt.title('Unemployement vs Date')
mpyplt.show()
```
